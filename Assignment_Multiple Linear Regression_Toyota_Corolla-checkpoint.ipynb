```python
# Price is Y
# Other columns are input variables X

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import numpy as np
```

# Preparing a model

```python
#Build model
import statsmodels.formula.api as smf 
model = smf.ols('Price~Age_08_04+KM+HP+Doors+Cylinders+Gears+Weight',data=cars).fit()
```

```python
#Coefficients
model.params
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)
```

```python
# R squared values
(model.rsquared,model.rsquared_adj)
```

```python
model.summary()
```

# Calculating VIF

```python
rsq_a = smf.ols('Age_08_04~KM+HP+Doors+Cylinders+Gears+Weight',data=cars).fit().rsquared
vif_a = 1/(1-rsq_a)

rsq_k = smf.ols('KM~Age_08_04+HP+Doors+Cylinders+Gears+Weight',data=cars).fit().rsquared
vif_k = 1/(1-rsq_k)

rsq_h = smf.ols('HP~Age_08_04+KM+Doors+Cylinders+Gears+Weight',data=cars).fit().rsquared
vif_h = 1/(1-rsq_h)

rsq_d = smf.ols('Doors~Age_08_04+KM+HP+Cylinders+Gears+Weight',data=cars).fit().rsquared
vif_d = 1/(1-rsq_d)

rsq_c = smf.ols('Cylinders~Age_08_04+KM+HP+Doors+Gears+Weight',data=cars).fit().rsquared
vif_c = 1/(1-rsq_c)

rsq_g = smf.ols('Gears~Age_08_04+KM+HP+Doors+Cylinders+Weight',data=cars).fit().rsquared
vif_g = 1/(1-rsq_g)

rsq_w = smf.ols('Weight~Age_08_04+KM+HP+Doors+Cylinders+Gears',data=cars).fit().rsquared
vif_w = 1/(1-rsq_w)

# Storing vif values in a data frame
d1 = {'Variables':['AGE','KM','HP','DOORS','CYL','GEARS','WEIGHT'],'VIF':[vif_a,vif_k,vif_h,vif_d,vif_c,vif_g,vif_w]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

# Test for Normality of Residuals (Q-Q Plot)

```python
import statsmodels.api as sm
qqplot = sm.qqplot(model.resid,line='q')  #Line = 45 degrees to the diagonal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid>6000))
```

# Residual Plot for Homoscedasticity

```python
def get_standardized_values(vals):return(vals-vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
           get_standardized_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

# Detecting Influencers/Outliers

# Cook’s Distance

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(cars)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#index and value of influencer where c is more than .5
(np.argmax(c),np.max(c))
```
